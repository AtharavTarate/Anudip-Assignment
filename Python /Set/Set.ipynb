Question 4: Suppose you have a set my_set = {1, 2, 3, {4, 5, 6}}, how can you access the nested set {4, 5, 6} from it? Write a Python program to achieve this.




```python
my_set = {1, 2, 3, frozenset({4, 5, 6})}
for element in my_set:
  if isinstance(element,frozenset):  # Check if the element is a set
    nested_set = element
    break  # Exit the loop once the nested set is found

print(nested_set)
```

```python
my_set = {1, 2, 3,(4, 5, 6)}
for element in my_set:
  if isinstance(element,tuple):  # Check if the element is a set
    nested_set = element
    break  # Exit the loop once the nested set is found

print(nested_set)
```
